Load Data

```python
import pandas as pd

data = pd.read_csv('../Dataset/smartphone_obsolescence.csv')
data
```

```python
import sklearn
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
data['Display_Tech'] = label.fit_transform(data['Display_Tech'])
data['Connectivity'] = label.fit_transform(data['Connectivity'])


custom_mapping = {
    'Best': 3,
    'Good': 2,
    'Below Average': 1,
    'Worst': 0
}

# Apply the mapping
data['Obsolescence_Category'] = data['Obsolescence_Category'].map(custom_mapping)
data
```

Multiple Linear Regression

```python
x = data[['Processor_Speed_GHz', 'RAM_GB', 'Storage_GB', 'Battery_mAh', 'Display_Tech', 'Connectivity', 'Camera_Score', 'Update_Support_Years', 'Security_Updates_Per_Year', 'Brand_Score', 'Release_Year', 'Initial_Price_USD', 'Customer_Rating']]
y = data['Obsolescence_Time_Years']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

from sklearn.metrics import mean_squared_error, r2_score
error = mean_squared_error(y_test, y_pred)
print(f"error = {error:.6f}")
r2 = r2_score(y_test, y_pred)
print(r2)

metrics = ['Mean Squared Error', 'R2 Score']
values = [error, r2]

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
bars = plt.bar(metrics, values, color = 'steelblue')

for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height(),
             f"{val:.4f}",
             ha='center', va='bottom')

plt.title("Model Performance Metrics for Multiple Linear Regression")
plt.ylabel("Metric Values")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

kNN Regressor

```python
x = data[['Processor_Speed_GHz', 'RAM_GB', 'Storage_GB', 'Battery_mAh', 'Display_Tech', 'Connectivity', 'Camera_Score', 'Update_Support_Years', 'Security_Updates_Per_Year', 'Brand_Score', 'Release_Year', 'Initial_Price_USD', 'Customer_Rating']]
y = data['Obsolescence_Time_Years']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors = 5)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

from sklearn.metrics import mean_squared_error, r2_score
error = mean_squared_error(y_test, y_pred)
print(f"error = {error:.6f}")
r2 = r2_score(y_test, y_pred)
print(r2)


metrics = ['Mean Squared Error', 'R2 Score']
values = [error, r2]

plt.figure(figsize=(8, 5))
bars = plt.bar(metrics, values, color = 'steelblue')

for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height(),
             f"{val:.4f}",
             ha='center', va='bottom')

plt.title("Model Performance Metrics for kNN Regressor")
plt.ylabel("Metric Values")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

SVM Regressor

```python
x = data[['Processor_Speed_GHz', 'RAM_GB', 'Storage_GB', 'Battery_mAh', 'Display_Tech', 'Connectivity', 'Camera_Score', 'Update_Support_Years', 'Security_Updates_Per_Year', 'Brand_Score', 'Release_Year', 'Initial_Price_USD', 'Customer_Rating']]
y = data['Obsolescence_Time_Years']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

from sklearn.svm import SVR
model = SVR(kernel = 'poly')
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

from sklearn.metrics import mean_squared_error, r2_score
error = mean_squared_error(y_test, y_pred)
print(f"Error = {error:.8f}")
r2 = r2_score(y_test, y_pred)
print(f"r2_score = {r2}")

metrics = ['Mean Squared Error', 'R2 Score']
values = [error, r2]

plt.figure(figsize=(8, 5))
bars = plt.bar(metrics, values, color = 'steelblue')

for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width() / 2,
             bar.get_height(),
             f"{val:.4f}",
             ha='center', va='bottom')

plt.title("Model Performance Metrics for SVM Regressor")
plt.ylabel("Metric Values")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
